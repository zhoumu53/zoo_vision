```python
import numpy as np
import rerun as rr

rr.init("test", spawn=True)

count = 5
rects_xy = np.random.rand(count, 2) * 1024
rects_uv = np.random.rand(count, 2) * 1024
rects = np.hstack((rects_xy, rects_uv))

rr.log(
    "/world",
    rr.Boxes2D(
        array=rects + 10,
        array_format=rr.Box2DFormat.XYXY,
        colors=np.asarray([255, 255, 0], dtype=np.uint8),
        labels=["in world"] * count,
    ),
)

rr.log("/world/boxes", rr.Transform3D(scale=0.5))
rr.log(
    "/world/boxes",
    rr.Boxes2D(
        array=rects,
        array_format=rr.Box2DFormat.XYWH,
        colors=np.asarray(
            [
                255,
                0,
                0,
            ],
            dtype=np.uint8,
        ),
        labels=["in world/boxes"] * count,
    ),
)  # Log the rectangles using Rerun
```
